## Creando el conjunto de datos

```python
import pandas as pd

# Lee el archivo CSV utilizando pandas
dataset1 = pd.read_excel('DB4.xlsx')

# Muestra las primeras filas del dataset
print(dataset1.head())
dataset1.shape
```

# Extrayendo columnas con Nulos

```python
cols_con_na=[col for col in dataset1.columns if (dataset1[col].isnull().mean()>0)]
cols_con_na
```

```python
prop_nan_per_col=dataset1[cols_con_na].isnull().mean()
prop_nan_per_col
```

Se puede determinar que las columnas, zona_ocu, Marca, Color y Modelo tienen un alto porcentaje de NA

```python
prop_nan_per_col2=dataset1[cols_con_na].median()
prop_nan_per_col2
```

Se cuentan con 4 columnas numericas, de las cuales 1 es continua y las demas son discretas
